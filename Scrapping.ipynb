```python
artist_names = []
for year in range(1998,2021,1):
    url = f'https://www.billboard.com/charts/year-end/{year}/top-r-and-b-hip-hop-artists'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    artists = soup.find_all("div", class_="ye-chart-item__title")
    for name in artists:
        names = name.find("a")
        artist_names.append(names.text.strip())
```
